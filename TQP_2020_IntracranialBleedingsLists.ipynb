```python
print("Number of subarachnoid bleeding patients before removing mixed bleedings: ", len(subarachnoid))

isolated_subarachnoid = set(subarachnoid).difference(intraparenchymal)
isolated_subarachnoid = set(isolated_subarachnoid).difference(epidural)
isolated_subarachnoid = set(isolated_subarachnoid).difference(subdural)

print("Number of isolated subarachnoid bleeding patients after removing mixed bleedings: ", len(isolated_subarachnoid))
```
