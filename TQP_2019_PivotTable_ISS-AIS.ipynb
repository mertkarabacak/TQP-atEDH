```python
import pandas as pd
import numpy as np
```

```python
#Read AIS lookup.

lookup = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_AISDIAGNOSIS_LOOKUP.csv", index_col = None, low_memory = False)
```

```python
ISSREGION = dict(zip(lookup.AISPREDOT, lookup.ISSREGION))
```

```python
#Read AIS data.

data = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_AISDIAGNOSIS.csv", index_col = None, low_memory = False)
```

```python
data['ISSREGION'] = data['AISPreDot'].map(ISSREGION)
```

```python
agg_functions = {'AISVersion': 'first', 'AISSeverity': 'max'}

iss_1 = data.loc[(data['ISSREGION'] == 1)]
iss_1 = iss_1.groupby(iss_1['inc_key']).aggregate(agg_functions)
#iss_1['ISSREGION1'] = 1
iss_1.rename(columns={"AISSeverity": "AISSeverity1"}, inplace=True)
iss_1['AISSeverity2'] = 0
iss_1['AISSeverity3'] = 0
iss_1['AISSeverity4'] = 0
iss_1['AISSeverity5'] = 0
iss_1['AISSeverity6'] = 0

iss_2 = data.loc[(data['ISSREGION'] == 2)]
iss_2 = iss_2.groupby(iss_2['inc_key']).aggregate(agg_functions)
#iss_2['ISSREGION2'] = 2
iss_2.rename(columns={"AISSeverity": "AISSeverity2"}, inplace=True)
iss_2['AISSeverity1'] = 0
iss_2['AISSeverity3'] = 0
iss_2['AISSeverity4'] = 0
iss_2['AISSeverity5'] = 0
iss_2['AISSeverity6'] = 0

iss_3 = data.loc[(data['ISSREGION'] == 3)]
iss_3 = iss_3.groupby(iss_3['inc_key']).aggregate(agg_functions)
#iss_3['ISSREGION3'] = 3
iss_3.rename(columns={"AISSeverity": "AISSeverity3"}, inplace=True)
iss_3['AISSeverity1'] = 0
iss_3['AISSeverity2'] = 0
iss_3['AISSeverity4'] = 0
iss_3['AISSeverity5'] = 0
iss_3['AISSeverity6'] = 0

iss_4 = data.loc[(data['ISSREGION'] == 4)]
iss_4 = iss_4.groupby(iss_4['inc_key']).aggregate(agg_functions)
#iss_4['ISSREGION4'] = 4
iss_4.rename(columns={"AISSeverity": "AISSeverity4"}, inplace=True)
iss_4['AISSeverity1'] = 0
iss_4['AISSeverity2'] = 0
iss_4['AISSeverity3'] = 0
iss_4['AISSeverity5'] = 0
iss_4['AISSeverity6'] = 0

iss_5 = data.loc[(data['ISSREGION'] == 5)]
iss_5 = iss_5.groupby(iss_5['inc_key']).aggregate(agg_functions)
#iss_5['ISSREGION5'] = 5
iss_5.rename(columns={"AISSeverity": "AISSeverity5"}, inplace=True)
iss_5['AISSeverity1'] = 0
iss_5['AISSeverity2'] = 0
iss_5['AISSeverity3'] = 0
iss_5['AISSeverity4'] = 0
iss_5['AISSeverity6'] = 0

iss_6 = data.loc[(data['ISSREGION'] == 6)]
iss_6 = iss_6.groupby(iss_6['inc_key']).aggregate(agg_functions)
#iss_6['ISSREGION6'] = 6
iss_6.rename(columns={"AISSeverity": "AISSeverity6"}, inplace=True)
iss_6['AISSeverity1'] = 0
iss_6['AISSeverity2'] = 0
iss_6['AISSeverity3'] = 0
iss_6['AISSeverity4'] = 0
iss_6['AISSeverity5'] = 0
```

```python
iss = pd.concat([iss_1, iss_2, iss_3, iss_4, iss_5, iss_6])
iss['inc_key'] = iss.index
```

```python
agg_functions = {'AISVersion': 'first', 'AISSeverity1': 'max', 'AISSeverity2': 'max', 'AISSeverity3': 'max', 'AISSeverity4': 'max', 'AISSeverity5': 'max', 'AISSeverity6': 'max'}
iss = iss.groupby(iss['inc_key']).aggregate(agg_functions)
```

```python
#Save pivot table for AIS Scores.

iss.to_csv('/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_AISDIAGNOSIS_PIVOT.csv')
```
