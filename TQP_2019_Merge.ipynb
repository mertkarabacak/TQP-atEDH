```python
import pandas as pd
import numpy as np
```

```python
#Read all data.

main_puf = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_TRAUMA.csv", index_col = 'inc_key', low_memory = False)
preexisting_conditions = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_PREEXISTINGCONDITIONS_PIVOT.csv", index_col = 'Inc_Key', low_memory = False)
hospital_events = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_HOSPITALEVENTS_PIVOT.csv", index_col = 'Inc_Key', low_memory = False)
ais = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_AISDIAGNOSIS_PIVOT.csv", index_col = 'inc_key', low_memory = False)
ecode = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_ECODE.csv", index_col = 'inc_key', low_memory = False)
interventions = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/PUF_INTERVENTIONS.csv", index_col = 'Inc_Key', low_memory = False)
isolated_intraparenchymal_ids = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_intraparenchymal_ids.csv", index_col = 0, low_memory = False)
isolated_epidural_ids = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_epidural_ids.csv", index_col = 0, low_memory = False)
isolated_subdural_ids = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_subdural_ids.csv", index_col = 0, low_memory = False)
isolated_subarachnoid_ids = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_subarachnoid_ids.csv", index_col = 0, low_memory = False)
```

```python
data = pd.concat([main_puf, preexisting_conditions, hospital_events, ais, ecode, interventions], axis=1)

data.shape
```

```python
isolated_intraparenchymal_ids = list(isolated_intraparenchymal_ids.iloc[:,0])
isolated_epidural_ids = list(isolated_epidural_ids.iloc[:,0])
isolated_subdural_ids = list(isolated_subdural_ids.iloc[:,0])
isolated_subarachnoid_ids = list(isolated_subarachnoid_ids.iloc[:,0])
```

```python
isolated_intraparenchymal = data.filter(items = isolated_intraparenchymal_ids, axis=0)

isolated_intraparenchymal.shape
```

```python
isolated_epidural = data.filter(items = isolated_epidural_ids, axis=0)

isolated_epidural.shape
```

```python
isolated_subdural = data.filter(items = isolated_subdural_ids, axis=0)

isolated_subdural.shape
```

```python
isolated_subarachnoid = data.filter(items = isolated_subarachnoid_ids, axis=0)

isolated_subarachnoid.shape
```

```python
#Save data frames.

isolated_intraparenchymal = pd.DataFrame(isolated_intraparenchymal)
isolated_intraparenchymal.to_csv('/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_intraparenchymal.csv')

isolated_epidural = pd.DataFrame(isolated_epidural)
isolated_epidural.to_csv('/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_epidural.csv')

isolated_subdural = pd.DataFrame(isolated_subdural)
isolated_subdural.to_csv('/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_subdural.csv')

isolated_subarachnoid = pd.DataFrame(isolated_subarachnoid)
isolated_subarachnoid.to_csv('/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_subarachnoid.csv')
```
