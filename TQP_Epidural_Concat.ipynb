```python
import pandas as pd
```

```python
#Read data dictionary.

data_dictionary = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/Modified Data Dictionary (Selected).csv", index_col = None, encoding = 'latin1', low_memory = False)

data_dictionary.columns = map(str.upper, data_dictionary.columns)
```

```python
variables = list(data_dictionary['VARIABLE'])

variables.remove('RACE')
variables.remove('PROTDEV')
variables.remove('ICP')
```

```python
#Read 2021 data.

df2021 = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2021/isolated_epidural.csv", index_col = 0, encoding = 'latin1', low_memory = False)

df2021.columns = map(str.upper, df2021.columns)
```

```python
#Remove unwanted columns from 2021 data.

df2021 = df2021[variables]
```

```python
#Check 2021 data shape.

df2021.shape
```

```python
#Read 2020 data.

df2020 = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2020/isolated_epidural.csv", index_col = 0, encoding = 'latin1', low_memory = False)

df2020.columns = map(str.upper, df2020.columns)
```

```python
#Remove unwanted columns from 2020 data.

df2020 = df2020[variables]
```

```python
#Check 2020 data shape.

df2020.shape
```

```python
#Read 2019 data.

df2019 = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2019/isolated_epidural.csv", index_col = 0, encoding = 'latin1', low_memory = False)

df2019.columns = map(str.upper, df2019.columns)
```

```python
#Change column name 'ISS_05' to 'ISS'.

df2019['ISS'] = df2019['ISS_05']
```

```python
#Remove unwanted columns from 2019 data.

df2019 = df2019[variables]
```

```python
#Check 2019 data shape.

df2019.shape
```

```python
#Combine data.

dfs = [df2019, df2020, df2021]

combined = pd.concat(dfs)
```

```python
#Check data shape.

combined.shape
```
