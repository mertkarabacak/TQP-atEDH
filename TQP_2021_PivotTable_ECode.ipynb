```python
import pandas as pd
import numpy as np
```

```python
#Read lookup.

lookup = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2021/PUF_ECODE_LOOKUP.csv", index_col = None, low_memory = False)
```

```python
INTENT = dict(zip(lookup.ECode, lookup.INTENT))
MECHANISM = dict(zip(lookup.ECode, lookup.MECHANISM))
TRAUMATYPE = dict(zip(lookup.ECode, lookup.TRAUMATYPE))
```

```python
#Read data.

data = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2021/PUF_TRAUMA.csv", index_col = None, low_memory = False)
```

```python
data = data[['inc_key', 'PRIMARYECODEICD10']]
data = data.set_index(['inc_key'])
```

```python
data['INTENT'] = data['PRIMARYECODEICD10'].map(INTENT)
data['MECHANISM'] = data['PRIMARYECODEICD10'].map(MECHANISM)
data['TRAUMATYPE'] = data['PRIMARYECODEICD10'].map(TRAUMATYPE)
```
