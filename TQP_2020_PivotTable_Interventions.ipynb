```python
import pandas as pd
import numpy as np
```

```python
#Read lookup.

lookup = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/Cranial Inverventions.csv", index_col = None, low_memory = False)
```

```python
cranial_interventions = list(lookup['ICD Code'])
```

```python
#Read ICD procedure data.

data = pd.read_csv("/content/drive/MyDrive/NTDB-PUFs/NTDB_2020/PUF_ICDPROCEDURE.csv", index_col = None, low_memory = False)
```

```python
def intervention(icd_p):
    if icd_p in cranial_interventions:
        return 1
    else:
        return np.NaN
```

```python
data['INTERVENTION'] = [intervention(x) for x in data['ICDPROCEDURECODE']]
```

```python
agg_functions = {'INTERVENTION': 'sum'}

pivot = data.groupby(data['Inc_Key']).aggregate(agg_functions)

pivot['INTERVENTION'].value_counts(normalize=False, dropna=False)
```

```python
#Change response values to strings for 'INTERVENTION'.

pivot.loc[pivot['INTERVENTION'] > 0, 'INTERVENTION'] = 'Yes'
pivot.loc[pivot['INTERVENTION'] == 0, 'INTERVENTION'] = 'No'

pivot['INTERVENTION'].value_counts(normalize=False, dropna=False)
```

```python
#Save data.

pivot.to_csv('/content/drive/MyDrive/NTDB-PUFs/NTDB_2020/PUF_INTERVENTIONS.csv')
```
